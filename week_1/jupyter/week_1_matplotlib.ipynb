```python
import matplotlib.pyplot as plt

from data_task01 import year, pop, life_exp, gdp_cap

# Print last year and last population
print("Last year with estimation:", year[-1])
print("Last estimated population:", pop[-1])

#First chart: population over time
plt.figure()
plt.plot(year, pop)
plt.xlabel("Year")
plt.ylabel("World Population [in billions]")
plt.title("World Population Projections 1950–2100")
plt.grid(True)

# Comment for first chart:
# According to the plot, the world population exceeds 10 billion
# slightly after 2080 (in the early 2080s).

plt.show()

#Second chart: GDP vs life expectancy
plt.figure()
plt.scatter(gdp_cap, life_exp)
plt.xlabel("GDP per Capita [in USD]")
plt.ylabel("Life Expectancy [in years]")
plt.title("World Development in 2007")
plt.xscale("log")

# Comment for second chart:
# The scatterplot shows a positive relationship: countries with higher GDP
# per capita tend to have higher life expectancy, with returns diminishing
# at very high income levels.

plt.show()
```

**Task 03**

```python
import matplotlib.pyplot as plt

from data_task03 import life_exp2007, life_exp1950

#1) Histogram of life expectancy in 2007
plt.figure()
plt.hist(life_exp, bins=15, edgecolor="black")
plt.xlabel("Life Expectancy [in years]")
plt.ylabel("Number of countries")
plt.title("Distribution of Life Expectancy in 2007")

# Comment / analysis of first plot:
# The 2007 life expectancy distribution is concentrated between about 50 and 80
# years, with most countries clustered in the 60–75 range and relatively few
# countries at very low life expectancy.

plt.show()

#2) Comparison: 1950 vs 2007, shared y-axis
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10, 4))

ax1.hist(life_exp1950, bins=15, edgecolor="black")
ax1.set_xlabel("Life Expectancy [in years]")
ax1.set_ylabel("Number of countries")
ax1.set_title("Life Expectancy in 1950")

ax2.hist(life_exp, bins=15, edgecolor="black")
ax2.set_xlabel("Life Expectancy [in years]")
ax2.set_title("Life Expectancy in 2007")

fig.suptitle("Global Life Expectancy: 1950 vs 2007")
fig.tight_layout()

# Comment / differences between the two expectancies:
# Compared to 1950, life expectancy in 2007 is clearly higher and less spread
# out: the whole distribution shifts to the right (fewer countries below ~50
# years, many more above 70), indicating substantial global improvements in
# health and longevity over time.

plt.show()
```

**Task 04**

```python
#recreating the plot by passing GDP per capita to x, life expectancy to y, and (doubled) population to the s argument of plt.scatter.

import numpy as np
import matplotlib.pyplot as plt

from data_task04 import gdp_cap, life_exp, pop

gdp_cap = np.array(gdp_cap)
life_exp = np.array(life_exp)
pop = np.array(pop)
sizes = pop * 2

plt.figure()
plt.scatter(gdp_cap, life_exp, s=sizes)
plt.xscale("log")
plt.xlabel("GDP per Capita [in USD]")
plt.ylabel("Life Expectancy [in years]")
plt.title("World Development in 2007")

plt.show()
```
