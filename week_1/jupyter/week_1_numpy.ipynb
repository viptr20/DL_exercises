**Task 2**

```python
import sys, os
sys.path.append("/content/sample_data")

import numpy as np
from baseball_dataset import baseball_dataset
from baseball_dataset_corrected import baseball_dataset_corrected


def main() -> None:
    np_baseball = np.array(baseball_dataset, dtype=float)

    print(f"Number of rows and columns: {np_baseball.shape}")

    height = np_baseball[:, 0]
    weight = np_baseball[:, 1]
    age = np_baseball[:, 2]

    print("Summary statistics for height:")
    print(f"  min: {height.min():.2f}, max: {height.max():.2f}, mean: {height.mean():.2f}")
    print("Summary statistics for weight:")
    print(f"  min: {weight.min():.2f}, max: {weight.max():.2f}, mean: {weight.mean():.2f}")
    print("Summary statistics for age:")
    print(f"  min: {age.min():.2f}, max: {age.max():.2f}, mean: {age.mean():.2f}")

    # Q1 answer:
    # In the raw data, the height column has impossible values like 74000.0,
    # 73000.0, 72000.0, 75000.0 inches, so the MLB data is clearly abnormal.

    # ---- CORRECTED DATA ----
    np_baseball_corr = np.array(baseball_dataset_corrected, dtype=float)
    h_corr = np_baseball_corr[:, 0]
    w_corr = np_baseball_corr[:, 1]
    a_corr = np_baseball_corr[:, 2]

    print("\nAfter data correction:")
    print("Summary statistics for height:")
    print(f"  min: {h_corr.min():.2f}, max: {h_corr.max():.2f}, mean: {h_corr.mean():.2f}")
    print("Summary statistics for weight:")
    print(f"  min: {w_corr.min():.2f}, max: {w_corr.max():.2f}, mean: {w_corr.mean():.2f}")
    print("Summary statistics for age:")
    print(f"  min: {a_corr.min():.2f}, max: {a_corr.max():.2f}, mean: {a_corr.mean():.2f}")

    # Extra statistics to answer: "Do big players tend to be heavier?"
    corr_hw = np.corrcoef(h_corr, w_corr)[0, 1]
    print("\nHeight–weight relationship (corrected data):")
    print(f"  Pearson correlation between height and weight: {corr_hw:.3f}")

    median_height = np.median(h_corr)
    taller = h_corr >= median_height
    shorter = ~taller

    print(f"  Mean weight for shorter half: {w_corr[shorter].mean():.2f}")
    print(f"  Mean weight for taller  half: {w_corr[taller].mean():.2f}")

    # Q2 answer:
    # The correlation between height and weight is clearly positive, and
    # the taller half of players has a higher mean weight than the shorter half.
    # So yes, bigger players do tend to be heavier.


if __name__ == "__main__":
    main()
```

**Task 4: Python vs NumPy**

```python
import numpy as np

# 1) array of zeros
py_zeros = [0] * 10
np_zeros = np.zeros(10)
print("Python:", py_zeros)
print("Numpy:", np_zeros)

# 2) array of ones
py_ones = [1] * 10
np_ones = np.ones(10)
print("Python:", py_ones)
print("Numpy:", np_ones)

# 3) array of fives
py_fives = [5] * 10
np_fives = np.full(10, 5, dtype=float)
print("Python:", py_fives)
print("Numpy:", np_fives)

# 4) array of zeroes, the fifth value of which is 1
py_zeros_fifth_one = [0] * 10
py_zeros_fifth_one[4] = 1

np_zeros_fifth_one = np.zeros(10, dtype=float)
np_zeros_fifth_one[4] = 1
print("Python:", py_zeros_fifth_one)
print("Numpy:", np_zeros_fifth_one)

# 5) array of all the integers from 10 to 50
py_10_to_50 = list(range(10, 51))
np_10_to_50 = np.arange(10, 51)
print("Python:", py_10_to_50)
print("Numpy:", np_10_to_50)

# 6) array of all the even integers from 10 to 50
py_even_10_to_50 = list(range(10, 51, 2))
np_even_10_to_50 = np.arange(10, 51, 2)
print("Python:", py_even_10_to_50)
print("Numpy:", np_even_10_to_50)

# 7) array that is the reversal of the list [1, 2, 3]
py_list = [1, 2, 3]
py_rev_slice = py_list[::-1]
py_rev_method = list(reversed(py_list))

np_arr = np.array([1, 2, 3])
np_rev_slice = np_arr[::-1]
np_rev_func = np.flip(np_arr)
print("Python:", py_rev_method)
print("Numpy:", np_rev_func)

# 8) a 3x3 matrix with values ranging from 0 to 8
py_matrix_3x3 = [[0, 1, 2],
                 [3, 4, 5],
                 [6, 7, 8]]
np_matrix_3x3 = np.arange(9).reshape(3, 3)
print("Python:", py_matrix_3x3)
print("Numpy:", np_matrix_3x3)

# 9) a 3x3 identity matrix
py_identity_3x3 = [[1 if i == j else 0 for j in range(3)] for i in range(3)]
np_identity_3x3 = np.eye(3, dtype=int)
print("Python:", py_identity_3x3)
print("Numpy:", np_identity_3x3)

# 10) a 5x5 matrix with values 1,2,3,4 just below the diagonal
py_5x5 = [[0]*5 for _ in range(5)]
for i, val in enumerate([1, 2, 3, 4], start=1):
    py_5x5[i][i-1] = val

np_5x5 = np.zeros((5, 5), dtype=int)
np.fill_diagonal(np_5x5[1:, :-1], [1, 2, 3, 4])
print("Python:", py_5x5)
print("Numpy:", np_5x5)
```

**Task 6: Use NumPy to generate:**

```python
import numpy as np

# 1) random number between 0 and 1
r1 = np.random.rand()
print("Random in [0,1]:", r1)

# 2) array of 25 random numbers from standard normal distribution
r2 = np.random.randn(25)
print("\n25 standard normal samples:\n", r2)

# 3) 3x3x3 array with random values from uniform [0,1)
r3 = np.random.rand(3, 3, 3)
print("\n3x3x3 uniform samples:\n", r3)

# 4) 10x10 array with uniform values and its min/max
r4 = np.random.rand(10, 10)
print("\n10x10 uniform array min:", r4.min(), "max:", r4.max())

# 5) random array (size 30, uniform) and its mean
r5 = np.random.rand(30)
print("\nSize-30 uniform array mean:", r5.mean())

# 6) multiply a 5x3 matrix by a 3x2 matrix
m1 = np.random.rand(5, 3)
m2 = np.random.rand(3, 2)
product = m1 @ m2   # or: np.dot(m1, m2)
print("\n5x3 @ 3x2 result shape:", product.shape)
print(product)
```

**Task 9: Let's practice composing operations:**

```python
import numpy as np

# 1) 1D array of random ints between -10 and 10; negate those in [3, 8]
arr1 = np.random.randint(-10, 11, size=10)
arr1[(arr1 >= 3) & (arr1 <= 8)] *= -1
print(arr1)

# 2) random float array of size 10, sorted
arr2 = np.random.rand(10)
arr2_sorted = np.sort(arr2)
print(arr2_sorted)

# 3) random uniform vector of size 10, replace max with 0
vec = np.random.rand(10)
print("The array:", vec)
vec[vec.argmax()] = 0.0
print("Result:", vec)
```
